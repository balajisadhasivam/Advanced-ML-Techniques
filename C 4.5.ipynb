```python
from chefboost import Chefboost as chef
```

```python
import pandas as pd

df = pd.read_csv("golf2.txt")
```

```python
df
```

```python
config = {'algorithm': 'C4.5'}
model = chef.fit(df, config = config, target_label = 'Decision')
```

```python
df
```

```python

prediction = chef.predict(model, ['Sunny',85,85,'Weak'])
```
